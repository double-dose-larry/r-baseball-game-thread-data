```python
from textblob import TextBlob
import pandas as pd
import datetime, pytz
import json
import requests
```

```python
df = pd.read_csv("ws_g7_gdt.csv")
```

```python
def sentiment_of_comment(comment_body):
    blob = TextBlob(str(comment_body))
    return sum([ sentence.sentiment.polarity for sentence in blob.sentences])
```

```python
df["sentiment"] = df.body.apply(sentiment_of_comment)
```

```python
df.groupby(["fanbase","inning"]).sentiment.sum().unstack()
```

```python
sentiment = df.groupby(["simplified_flair", "inning"]).sentiment.sum().unstack()
```

```python
#fix order of columns
cols = sentiment.columns.tolist()
cols = cols[-1:] + cols[:-1]
sentiment = sentiment[cols]
```
